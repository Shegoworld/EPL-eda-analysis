```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path = '/kaggle/input/datasets/shegoworld/epl-stats/epl-stats2020.csv'
my_data = pd.read_csv(path)
#print(my_data.head())
```

## Top 10 EPL Scorers
Which players scored the most goals from the dataset?
Vardy and Aubameyang lead the chart, with strikers dominating the top 10 as expected.

```python
#Top 10 EPL Scorers
scorers = my_data.sort_values('Goals', ascending = False).head(10)
plt.figure(figsize = (10, 6))
plt.title("Top 10 Top Scorers")
sns.barplot(data = scorers, x = 'Name', y = 'Goals')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
```

**Top 10 EPL Scorers**

Which players scored the most goals in the dataset?
Vardy and Aubameyang lead the chart, with strikers dominating the top 10 as expected.

```python
#Bar chart — Average goals per position
avg_goals = my_data.groupby('Position')['Goals'].mean().reset_index()
plt.figure(figsize = (8,6))
plt.title("Average Goals by Position")
sns.barplot(data = avg_goals, x = 'Position', y = 'Goals')
plt.tight_layout()
```

**Average Goals by Position**

Forwards have the highest goals on average as shown in the image

```python
#Scatter plot — Appearances vs Goals (does more playing time mean more goals?)
plt.figure(figsize = (10,6))
plt.title("Relationship between Appearances and Goals")
sns.regplot(data = my_data, x = 'Appearances', y = 'Goals')
plt.tight_layout()
```

**Does Appearances affect goal scored?**

There is a positive relationship between appearances and goal scored which means players with more appearances tend to score more goals

```python
#Bar chart — Top 5 most efficient forwards
my_dataf = my_data[(my_data.Appearances > 0) & (my_data.Position == 'Forward')].copy() 
my_dataf["Goal Efficiency"] = my_dataf['Goals'] / my_dataf['Appearances']
my_dataf = my_dataf.sort_values('Goals', ascending = False).head()
plt.figure(figsize = (8,6))
plt.title("Top 5 Most Efficient Forwards")
sns.barplot(data = my_dataf, x = 'Name', y = 'Goal Efficiency')
plt.tight_layout()
```

**Who are the most efficient forwards?**

Aguero closely followed by Kane tops the list of the forwards goal efficiency

```python
#Total Goals by Club
club_goals = my_data.groupby('Club')['Goals'].sum().reset_index()
plt.figure(figsize = (10,6))
plt.title("Total Goals by Clubs")
sns.barplot(data = club_goals, x = 'Goals', y = 'Club')
plt.tight_layout()
```

**Which clubs have scored the most goals?**

Manchester City is leading in total goals scored by clubs
